```python
def plota_gastos_per_capita(dados): # uma função em python
  media = np.mean(dados)
  plt.bar(range(len(dados)), dados)
  plt.axhline(media, color='red', linestyle='dashed')
  plt.ylabel("Reais/habitante")
  plt.xlabel("Unidade da federação")
  plt.title("Gastos per capita por estado com media para o último mês")
  plt.show()
```

```python
def plota_gastos_per_capita2(dados): # uma função em python
  media = np.mean(dados)
  ax = dados.plot.bar(label="Gastos Per Capita em BRL")
  ax.axhline(media, color='red', linestyle='dashed', label='Média')
  ax.legend()
  ax.set_title('Gráfico de barras com os gastos per capita')
  ax.set_ylabel("Reais/habitante")
  ax.set_xlabel("Unidade da federação")
  plt.show()
```

```python
def insere_gastos_e_gastos_por_habitante(ordenados_por_total, gastos_e_populacao, mes):
  gastos = ordenados_por_total[mes]
  gastos_e_populacao[f"gastos_{mes}"] = gastos * 1_000_000
  gastos_e_populacao[f"{mes}_gastos_per_capita"] = gastos_e_populacao[f"gastos_{mes}"] / gastos_e_populacao["Pop_Estimada"]
  return gastos_e_populacao
```
